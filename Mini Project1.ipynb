### FIND MEDIAN FROM DAILY REVENUE

```python
# upload dataset
import pandas as pd
data = pd.read_csv('01 telecom_revenue.csv')

# filtering MonthlyRevenue yang tidak sama dengan 0
data_1 = data[data['MonthlyRevenue']!= 0]

# filtering TotalRecurringCharge yang tidak sama dengan 0
data_2 = data_1[data_1['TotalRecurringCharge'] != 0]
data_2
```

```python
# copy data_2
data_2_new = data_2.copy()
# make a new column
data_2_new['revenue per recurring charge'] = data_2_new['MonthlyRevenue']/data_2_new['TotalRecurringCharge']
data_2_new
```

```python
# dataset for revenue per recurring charge > 9
data_2_new_over_9 = data_2_new[~(data_2_new['revenue per recurring charge']>9)]
data_2_new_over_9
```

```python
# copy dataset
data_2_new_over_9_new = data_2_new_over_9.copy()
# make a column DailyRevenue
data_2_new_over_9_new['DailyRevenue'] = data_2_new_over_9_new['MonthlyRevenue']/30
data_2_new_over_9_new.head(5)
```

```python
# calculate median from DailyRevenue
result = data_2_new_over_9_new['DailyRevenue'].median()
print('median DailyRevenue =', result)
```
